```python
import pandas as pd

df = pd.read_csv("dataset1.csv")
df.head()
```

```python
correlation = df.corr(numeric_only=True)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.show()
```
